## 1. Build hypothesis and cost

$H(x) = Wx + b$ : 가설함수

### hypothesis

```python
# data
x_data = [1,2,3,4,5]
y_data = [1,2,3,4,5]

W = tf.Variable(2.9)
b = tf.Variable(0.5)

# typothesis = W * x + b
hypothesis = W * x_data + b
```

```python
W.numpy(), b.numpy()
```

```python
hypothesis.numpy()
```

```python
plt.plot(x_data, hypothesis, '-r')
plt.plot(x_data, y_data, 'o')
plt.ylim(0, 8)
plt.show()
```

### cost

$cost(W, b) = \frac{1}{m}\sum(H(x^{(i)} - y^{(i)})^2$

```python
v = [1., 2., 3., 4.]
print(tf.rank(v))

v1 = tf.reduce_mean(v)
print(v1)
print(tf.rank(v1))

print(tf.square(3))
```

```python
# 비용함수
cost = tf.reduce_mean(tf.square(hypothesis - y_data))
```

## 2. Gradient descent
경사 하강법

```python
# Learning rate intialize
learning_rate = 0.01

# Gradient descent
with tf.GradientTape() as tape:
    hypothesis = W * x_data + b
    cost = tf.reduce_mean(tf.square(hypothesis - y_data))
    
W_grad, b_grad = tape.gradient(cost, [W, b])

W_grad.numpy(), b_grad.numpy()
```

```python
# Parameter(W,b) Update
W.assign_sub(learning_rate * W_grad)
b.assign_sub(learning_rate * b_grad)

W.numpy(), b.numpy()
```

```python
plt.plot(x_data, hypothesis, '-r')
plt.plot(x_data, y_data, 'o')
plt.ylim(0, 8)
plt.show()
```

## 3. repeat several times

```python
W = tf.Variable(2.9)
b = tf.Variable(0.5)

for i in range(100):
    with tf.GradientTape() as tape:
        hypothesis = W * x_data + b
        cost = tf.reduce_mean(tf.square(hypothesis - y_data))
    W_grad, b_grad = tape.gradient(cost, [W, b])
    W.assign_sub(learning_rate * W_grad)
    b.assign_sub(learning_rate * b_grad)

    if i%10 == 0:
        print("{:5}|{:10.4f}|{:10.4f}|{:10.6f}".format(i, W.numpy(), b.numpy(), cost))
```

```python
plt.plot(x_data, y_data, 'o')
plt.plot(x_data, hypothesis.numpy(), 'r-')
plt.ylim(0, 8)
```

## 4. Predict

```python
print(W * 5 + b) # 기댓값 : 5
print(W * 2.5 + b) # 기댓값 : 2.5
```

# 전체 코드

```python
import tensorflow as tf
import numpy as np

# data
x_data = [1,2,3,4,5]
y_data = [1,2,3,4,5]

# W, b initialize
W = tf.Variable(2.9)
b = tf.Variable(0.5)

learning_rate = 0.01

for i in range(100):
    # Gradient descent
    with tf.GradientTape() as tape:
        hypothesis = W * x_data + b
        cost = tf.reduce_mean(tf.square(hypothesis - y_data))
    W_gred, b_gred = tape.gradient(cost, [W, b])
    W.assign_sub(learning_rate * W_gred)
    b.assign_sub(learning_rate * b_gred)
    
    if i%10 == 0:
        print(f"{i:5}|{W.numpy():10.4f}|{b.numpy():10.4f}|{cost:10.6f}")
        
# predict
print(W * 5 + b)
print(W * 2.5 + b)
```
